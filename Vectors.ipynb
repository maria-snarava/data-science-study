Here I use matplotlib to show vectors on the plot

```python
import time 
import sys
import numpy as np 
import matplotlib.pyplot as plt


def Plotvec3(a, b, sum_a_b, subtract, multiply_a_b, divide_b_a):
    ax = plt.axes()# to generate the full window axes
    
    #Add an arrow to the a Axes with arrow head width 0.05, color red and arrow head length 0.1
    ax.arrow(0, 0, *a, head_width=0.05, color ='r', head_length=0.1)
    plt.text(*(a + 0.1), 'a')
    
    #Add an arrow to the  b Axes with arrow head width 0.05, color blue and arrow head length 0.1
    ax.arrow(0, 0, *b, head_width=0.05, color ='r', head_length=0.1)
    plt.text(*(b + 0.1), 'b')

    ax.arrow(0, 0, *sum_a_b, head_width=0.05, color ='g', head_length=0.1)
    plt.text(*(sum_a_b + 0.1), 'sum_a_b')
    
    ax.arrow(0, 0, *subtract, head_width=0.05, color ='b', head_length=0.1)
    plt.text(*(subtract + 0.1), 'subtract')
    
    ax.arrow(0, 0, *multiply_a_b, head_width=0.05, color ='y', head_length=0.1)
    plt.text(*(multiply_a_b + 0.1), 'multiply_a_b')
    
    ax.arrow(0, 0, *divide_b_a, head_width=0.05, color ='k', head_length=0.1)
    plt.text(*(divide_b_a + 0.1), 'divide_b_a')
        
    plt.ylim(-0.5, 5)#set the ylim
    plt.xlim(-0.5, 3)#set the xlim

a = np.array([1, 0.5])
b = np.array([1, 2])

sum_a_b = a + b
print(f"The sum is {sum_a_b}")
subtract = b - a
print(f"The subtract is {subtract}")
multiply_a_b = a * b
print(f"The multiply_a_ is {subtract}")
divide_b_a = b / a
print(f"The divide is {divide_b_a}")

Plotvec3(a, b, sum_a_b, subtract, multiply_a_b, divide_b_a)
print("The dot product is", np.dot(a,b))
```
